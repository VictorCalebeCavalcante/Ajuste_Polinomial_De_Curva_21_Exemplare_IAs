# Divisão Dos Exempalres em Set de treinamento e Set de Teste:

```
import numpy as np

h = lambda x: -0.85*np.exp(0.9*x) + 0.36 + 0.12*np.sin(5*np.pi*x)

x = np.linspace(0,1,200)

xtr = x[0:134]

xte = x[134:200]

htr = h(xtr)

hte = h(xte)

```

# Geração de Valores Aleatórios de *Distribuição Gaussiana*:
Adição do "noise" aos valores de "xtr".

```
mu, sigma = 0, 0.05
noise = np.random.normal(mu,sigma,134)
xtrp = xtr + noise
xtrp
```

# Treinamento e Plotagem das Curvas:

```
%matplotlib inline
from numpy import linspace, zeros, pi, cos, sin, exp
from pylab import plot, grid, xlabel, ylabel, title, figure, subplot, tight_layout, axis, legend, text, xlim, ylim

peso1_tr = np.polyfit(xtrp,htr,1)#Pesos da curva_1
peso3_tr = np.polyfit(xtrp,htr,3)#Pesos da curva_3
peso5_tr = np.polyfit(xtrp,htr,5)#Pesos da curva_5
peso7_tr = np.polyfit(xtrp,htr,7)#Pesos da curva_7
peso9_tr = np.polyfit(xtrp,htr,9)#Pesos da curva_9
peso11_tr = np.polyfit(xtrp,htr,11)#Pesos da curva_11
peso13_tr = np.polyfit(xtrp,htr,13)#Pesos da curva_13


ajuste1_tr = np.polyval(peso1_tr,xtrp)#Cuva ajustada com os pontos
ajuste3_tr = np.polyval(peso3_tr,xtrp)#...
ajuste5_tr = np.polyval(peso5_tr,xtrp)#...
ajuste7_tr = np.polyval(peso7_tr,xtrp)#...
ajuste9_tr = np.polyval(peso9_tr,xtrp)#...
ajuste11_tr = np.polyval(peso11_tr,xtrp)#...
ajuste13_tr = np.polyval(peso13_tr,xtrp)#Cuva ajustada com os pontos


figure(figsize=(14,6))

plot(xtrp,h(xtrp),color = 'red') # h(x)

plot(xtrp, ajuste1_tr,'k:',color = 'blue') #Fitting
plot(xtrp, ajuste3_tr,'k:',color = 'green') #Fitting
plot(xtrp, ajuste5_tr,'k:', color = 'orange') #Fitting
plot(xtrp, ajuste7_tr, color = 'yellow') #Fitting
plot(xtrp, ajuste9_tr, color = 'pink') #Fitting
plot(xtrp, ajuste11_tr, color = 'brown') #Fitting
plot(xtrp, ajuste13_tr, color = 'black') #Fitting

#
grid(True)
xlabel('$x$',fontsize=16)
ylabel('$ajuste(x)$',fontsize=16)
title('Ajuste Polinomail')
legend(['h(x)','m = 1','m = 3', 'm = 5','m = 7', 'm = 9', 'm = 11', 'm = 13'])
xlim(xtrp[0],xtrp[-1])
ylim(-1.25,-0.25)

```




# Teste e Plotagem das Curvas:

```
ajuste1_te = np.polyval(peso1_tr,xte)
ajuste3_te = np.polyval(peso3_tr,xte)
ajuste5_te = np.polyval(peso5_tr,xte)
ajuste7_te = np.polyval(peso7_tr,xte)
ajuste9_te = np.polyval(peso9_tr,xte)
ajuste11_te = np.polyval(peso11_tr,xte)
ajuste13_te = np.polyval(peso13_tr,xte)

figure(figsize=(14,6))

plot(xte,h(xte),color = 'red') # correto

plot(xte, ajuste1_te,'k:',color = 'blue') #Fitting
plot(xte, ajuste3_te,'k:',color = 'green') #Fitting
plot(xte, ajuste5_te,'k:', color = 'orange') #Fitting
plot(xte, ajuste7_te, color = 'yellow') #Fitting
plot(xte, ajuste9_te, color = 'pink') #Fitting
plot(xte, ajuste11_te, color = 'brown') #Fitting
plot(xte, ajuste13_te, color = 'black') #Fitting

grid(True)
xlabel('$x$',fontsize=16)
ylabel('$ajuste(x)$',fontsize=16)
title('Ajuste Polinomial')
legend(['h(x)', 'm = 1', 'm = 3', 'm = 5', 'm = 7', 'm = 9'])
xlim(xte[0],xte[-1])
ylim(h(xte)[0]-1,h(xte)[-1]+1)
```

# Treino e Calculo do Erro :

```
#treino
import math

erms_tr_wx = []
m_tr = [1,3,5,7,9,11,13]

erms_w1 = (ajuste1_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w1)/134))

erms_w3 = (ajuste3_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w3)/134))

erms_w5 = (ajuste5_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w5)/134))

erms_w7 = (ajuste7_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w7)/134))

erms_w9 = (ajuste9_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w9)/134))

erms_w11 = (ajuste11_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w11)/134))

erms_w13 = (ajuste13_tr - h(xtr))**2
erms_tr_wx.append(math.sqrt(sum(erms_w13)/134))

plot(m_tr,erms_tr_wx)

grid(True)
xlabel('$m$',fontsize=16)
ylabel('$E^{RMS}(m)$',fontsize=16)
title('$E^{RMS} X Grau$')

print(erms_tr_wx)
```
